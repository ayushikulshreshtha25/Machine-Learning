```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("1.03. Dummies.csv")
data
```

```python
data1=data.copy()
```

```python
data1['Attendance']=data1['Attendance'].map({'Yes':1,'No':0})
```

```python
data1
```

```python
data1.describe()
```

```python
y=data1['GPA']
x=data1[['SAT','Attendance']]
```

```python
x1=sm.add_constant(x)
```

```python
result=sm.OLS(y,x1).fit()
result.summary()
```

```python
plt.scatter(data1['SAT'],y,c=data1['Attendance'],cmap='RdYlGn_r')
yhat_no=0.0014*data1['SAT']+0.6439
yhat_yes=0.0014*data1['SAT']+0.8665
yhat=0.0017*data1['SAT']+0.275
fig=plt.plot(data1['SAT'],yhat_no,lw=4,c='green',label='regression line')
fig=plt.plot(data1['SAT'],yhat_yes,lw=4,c='red',label='regression line')
fig=plt.plot(data1['SAT'],yhat,lw=4,c='blue',label='regression line')
plt.xlabel("SAT",fontsize=20)
plt.ylabel("GPA",fontsize=20)
plt.show()
```

```python
newdata=pd.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
newdata=newdata[['const','SAT','Attendance']]
newdata
```

```python
newdata.rename(index={0:'Bob',1:'Alice'})
```

```python
prediction=result.predict(newdata)
prediction
```
