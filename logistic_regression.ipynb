###sample of logistic model using indian diabetis data

```python
import pandas as pd
import numpy as np
import pickle as pi
import matplotlib as mt
import sklearn as si
import scipy as sc
```

```python
df = pd.read_csv('pima_indian_diabetes.csv')
df
```

```python
db = df.groupby(['Diabetes'])['Age'].mean()

db.plot.barh()
```

```python
shuffled_indices = np.random.permutation(len(df))
test_set_size = int(len(df) * 0.3)
test_indices = shuffled_indices[:test_set_size]
train_indices = shuffled_indices[test_set_size:]
```

```python
train=df.iloc[train_indices]
train
```

```python
train.mean()
```

```python
test=df.iloc[test_indices]
test
```

```python
test.mean()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(train.drop('Diabetes',axis=1),train['Diabetes'])
```

```python
predict=log_reg.predict(test.drop('Diabetes',axis=1))
predict
```

```python
from sklearn.metrics import confusion_matrix
ar=confusion_matrix(predict,test['Diabetes'])
ar
```
